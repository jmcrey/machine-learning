### Problem 1
In this question, we will use the csv file **Temp.csv**. 
- Read it into a data frame, call it `Temp`. (Import the packages you think you need).

```python
import pandas as pd
Temp = pd.read_csv("Temp.csv")
print(type(Temp))
Temp
```

- We learned that the temperature at New Haven, Connecticut is currently 58 degree fahrenheit. Add this information into the DataFrame.

```python
new_data = pd.DataFrame([["New Haven", "Connecticut", 58]], columns=["City", "State", "Temperature"], index=[max(Temp.index) + 1])
Temp = Temp.append(new_data)
Temp
```

- Replace the names of the states by their abbreviations. That is 
    - Connecticut --> CT
    - New Jersey  --> NJ
    - New York    --> NY

```python
state_abbr = dict(State={
    "New York": "NY",
    "Connecticut": "CT",
    "New Jersey": "NJ"
})
Temp.replace(to_replace=state_abbr, inplace=True)
Temp
```

- Select the rows in New York.

```python
new_york = Temp.loc[Temp['State'] == 'NY']
new_york
```

- What is the average of the temperature of cities in NY?

```python
avg_ny = sum(new_york['Temperature']) / len(new_york['Temperature'])
avg_ny
```

- Find the mean and the standard deviation of the temperature of each state.

```python
## TODO: find the mean and standard deviation of temp in each state
for state, df in Temp.groupby('State'):
    mean = sum(df['Temperature']) / len(df['Temperature'])
    stddev = sum(pow(df['Temperature'] - mean, 2)) / len(df['Temperature'])
    print(f"Mean temperature for {state} is {mean}, and the standard deviation is {stddev}")
```

- We know the capital of each state: 
    - Connecticut --> Hartford
    - New Jersey  --> Treton
    - New York    --> Albany
    
    Create a data frame, `Capital`, with two columns: State and Capital, which should look like:
    
|| State| Capital|
| :-: |:---: | :---:|
| 0| CT  | Hartford|
| 1| NJ | Treton|
| 2| NY| Albany|

```python
capital_data = [
    ["CT", "Hartford"],
    ["NJ", "Treton"],
    ["NY", "Albany"]
]
Capital = pd.DataFrame(capital_data, columns=["State", "Capital"])
Capital
```

- Mutate a new boolean column, `Capital`, to `Temp`, which is `True` if the city in the row is the capital of the state, else `False`.

```python
Temp['Capital'] = Temp.apply(lambda x: (
        Capital.loc[Capital['State'] == x['State']]['Capital'] == x['City']
    ).bool(), axis=1)
Temp
```

### Problem 2

- a. Load the Gross Domestic Product data for the 190 ranked countries in the file **GDP.csv**. Load **Country.csv** into a DataFrame. 
- b. Match the data based on the country shortcode. How many of the IDs match? 
- c. Sort the data frame in descending order by GDP rank. What is the 13th country in the resulting data frame? (Note: GDP formats as 123,456,789.)
- d. What is the average GDP for the "High income: OECD" and "High income: nonOECD" group?
- e. What is the average GDP for different income groups and regions?

```python
gdp = pd.read_csv("GDP.csv", encoding="ISO-8859-1", skiprows=[0,1,4], header=1, usecols=[0,1,3,4], nrows=190, index_col=0)
gdp
```

```python
country = pd.read_csv("Country.csv", encoding="ISO-8859-1", index_col=0)
country
```

```python
merged = country.merge(gdp, left_index=True, right_index=True)
print(f"{len(merged)} country codes match")
merged
```

```python
ranked = merged.sort_values('Ranking', ascending=False)
print(f"The 13th ranked country (descending order) is {ranked.iloc[12]['Long Name']}")
ranked.iloc[12]
```

```python
income_group = ranked.groupby('Income Group')
targets = ["High income: OECD", "High income: nonOECD"]
for group, df in income_group:
    if group in targets:
        mean = sum(int(gdp.replace(',', '')) for gdp in df['US dollars)']) / len(df['US dollars)'])
        print(f"Average GDP for group \"{group}\" is {mean}")
```

```python
income_regions = ranked.groupby(['Income Group', 'Region'])
max_group = max(len(group) for group in ranked['Income Group'])
max_region = max(len(region) for region in ranked['Region'])
for info, df in income_regions:
    group, region = info
    group = f"\"{group}\"{' ' * (max_group - len(group))}"
    region = f"\"{region}\"{' ' * (max_region - len(region))}"
    mean = sum(int(gdp.replace(',', '')) for gdp in df['US dollars)']) / len(df['US dollars)'])
    print(f"Group: {group}\tRegion: {region}\tAverage GDP: {mean}")
```
